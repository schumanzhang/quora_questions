### We will engineer some features here, namely three basic features:

#### 1) character length of questions 1 and 2
#### 2) number of words in questions 1 and 2
#### 3) normalized word share count


```python
import nltk

display(sub_sample.head())
```

```python
len(sub_sample['question1'][0])

question1_arr = sub_sample['question1'].values
question2_arr = sub_sample['question2'].values

print(question1_arr[:2])
print(question2_arr[:2])
```

```python
question1_char_length = [len(text) for text in question1_arr]
question2_char_length = [len(text) for text in question2_arr]

print(question1_char_length[:5])
print(question2_char_length[:5])
```

```python
labels = sub_sample['is_duplicate'].values

def char_length_diff(arr1, arr2, labels):
    char_diff = []
    for index, val in enumerate(arr1):
        diff = (abs(val - arr2[index]), labels[index])
        char_diff.append(diff)
    return char_diff

char_diff = char_length_diff(question1_char_length, question2_char_length, labels)

print(char_diff[:5])    
```

```python
col = np.where(labels > 0 ,'r','k')

y = [item[0] for item in char_diff]
x = [index + 1 for index, val in enumerate(char_diff)]

plt.scatter(x, y, label='char_length_diff', color=col)
plt.rcParams['figure.figsize'] = [20, 10]

plt.xlabel('Question-pair index')
plt.ylabel('Character differences')
plt.title('Question-pair character differences')
plt.legend()
plt.show()
```

We can see from the above duplicates tend to have shorter character differences between question pairs. We can extend the analysis to difference in word counts...

```python
# question1_arr
# question2_arr nltk.word_tokenize(sentence)
nltk.download('punkt')

question1_word_length = [len(nltk.word_tokenize(text)) for text in question1_arr]
question2_word_length = [len(nltk.word_tokenize(text)) for text in question2_arr]

print(question1_arr[0])
print(question1_word_length[0])
```

```python
def word_length_diff(arr1, arr2, labels):
    char_diff = []
    for index, val in enumerate(arr1):
        diff = (abs(val - arr2[index]), labels[index])
        char_diff.append(diff)
    return char_diff

word_diff = word_length_diff(question1_word_length, question2_word_length, labels)

print(word_diff[:5])  
```

```python
col = np.where(labels > 0 ,'r','k')

y = [item[0] for item in word_diff]
x = [index + 1 for index, val in enumerate(word_diff)]

plt.scatter(x, y, label='word_length_diff', color=col)
plt.rcParams['figure.figsize'] = [20, 10]

plt.xlabel('Question-pair index')
plt.ylabel('Word count differences')
plt.title('Question-pair character differences')
plt.legend()
plt.show()
```

We see a very similar pattern emerge here. We can visualize another way...

```python
# Calculate word-share coefficient
from nltk.corpus import stopwords
nltk.download('stopwords')

stops = set(stopwords.words("english"))
```

```python
# question1_arr
# question2_arr

def calc_wordshare(arr1, arr2):
    wordshare_coefficient = []
    for index, val in enumerate(arr1):
        q1_words = {}
        q2_words = {}
        q1words = nltk.word_tokenize(val)
        q2words = nltk.word_tokenize(arr2[index])
        
        for word in q1words:
            if str(word) not in stops:
                q1_words[str(word)] = 1
                
        for word2 in q2words:
            if str(word2) not in stops:
                q2_words[str(word2)] = 1
        
        shared_words_in_q1 = [w for w in q1_words.keys() if w in q2_words]
        shared_words_in_q2 = [w for w in q2_words.keys() if w in q1_words]
        
        R = len(shared_words_in_q1) + len(shared_words_in_q2) / (len(q1_words) + len(q2_words))
        
        wordshare_coefficient.append(R)

    return wordshare_coefficient
    
wordshare_coefficient = calc_wordshare(question1_arr, question2_arr)

print(wordshare_coefficient[:2])
print(len(wordshare_coefficient))
```

```python
col = np.where(labels > 0 ,'r','k')

y = wordshare_coefficient
x = [index + 1 for index, val in enumerate(wordshare_coefficient)]

plt.scatter(x, y, label='wordshare_coefficient', color=col)
plt.rcParams['figure.figsize'] = [20, 10]

plt.xlabel('Question-pair index')
plt.ylabel('wordshare_coefficient')
plt.title('Wordshare_coefficient feature')
plt.legend()
plt.show()
```

```python
# Improving wordshare coefficient with TF-IDF
vectorizer = TfidfVectorizer()

tf_idf_matrix = vectorizer.fit_transform(dfqa['question'].values)
idf = vectorizer.idf_
vocab_dict = dict(zip(vectorizer.get_feature_names(), idf))

print(vocab_dict.get('what'))
something = vectorizer.inverse_transform(tf_idf_matrix[0])
print(question2_arr[0])
```

```python
def tf_idf_wordshare(arr1, arr2, vocab_dict):
    tf_idf_wordshare_coefficient = []
    for index, val in enumerate(arr1):
        q1_words = {}
        q2_words = {}
        q1words = nltk.word_tokenize(val)
        q2words = nltk.word_tokenize(arr2[index])
        
        for word in q1words:
            if word.lower() not in stops:
                q1_words[word.lower()] = 1
                
        for word2 in q2words:
            if word2.lower() not in stops:
                q2_words[word2.lower()] = 1
                
        shared_weights = [vocab_dict.get(w) for w in q1_words.keys() if w in q2_words] + [vocab_dict.get(w) for w in q2_words.keys() if w in q1_words]
        total_weights = [vocab_dict.get(w) for w in q1_words] + [vocab_dict.get(w) for w in q2_words]
        
        R = np.sum([x for x in shared_weights if x is not None]) / np.sum([x for x in total_weights if x is not None])
        
        tf_idf_wordshare_coefficient.append(R)
    
    return tf_idf_wordshare_coefficient
    
tf_idf_wordshare_coefficient = tf_idf_wordshare(question1_arr, question2_arr, vocab_dict)

print(tf_idf_wordshare_coefficient[:5])
print(len(tf_idf_wordshare_coefficient))
```

```python
col = np.where(labels > 0 ,'r','k')

y = tf_idf_wordshare_coefficient
x = [index + 1 for index, val in enumerate(tf_idf_wordshare_coefficient)]

plt.scatter(x, y, label='tf_idf_wordshare_coefficient', color=col)
plt.rcParams['figure.figsize'] = [20, 10]

plt.xlabel('Question-pair index')
plt.ylabel('tf_idf_wordshare_coefficient')
plt.title('tf_idf_wordshare_coefficient feature')
plt.legend()
plt.show()
```
